```python
# get dates as (complicated) strings in a numpy array
das = temps['datetime'].values

# get temperatures for detroit in a simple numpy array
detroit_temp = temps['Detroit'].values
```

```python
das
```

```python
dates = pd.to_datetime(das,format='%Y-%m-%d %H:%M:%S')
```

```python
dates
```

```python
dates.year
```

```python
dates.month
```

```python
dates.day
```

```python
date_for_temp = []
temp_mean = []
temp_min = []
temp_max = []

for year in range(2012,2018):
    for month in range(1,13):

        # get this month's temperatures in Celsius
        this_month_temps = detroit_temp[ np.logical_and(np.logical_and(dates.year == year, dates.month == month), 
                                                np.isfinite(detroit_temp))] - 273

        if this_month_temps.size > 0:
            date_for_temp.append(datetime(year,month,15))
        
            temp_mean.append(this_month_temps.mean())
            temp_max.append(this_month_temps.max())
            temp_min.append(this_month_temps.min())
```

```python
date_for_temp
```

```python
plt.plot(dates,detroit_temp-273.0,'k-',alpha=0.2,zorder=1)

plt.fill_between(date_for_temp, temp_min, temp_max, facecolor='red', alpha=.3, zorder=2)

plt.plot(date_for_temp, temp_mean, 'b', linewidth=3,zorder=3)

plt.ylabel('Temperature (Celsius)')
plt.title('Detroit Temperature')
```

```python
plt.plot(dates,detroit_temp-273.0,'k-',alpha=0.2,zorder=1)

plt.fill_between(date_for_temp, temp_min, temp_max, facecolor='red', alpha=.3, zorder=2)

plt.plot(date_for_temp,temp_mean,'b',linewidth=3,zorder=3)

plt.xlabel('Date')
plt.ylabel('Temperature [C]')
plt.title('Detroit min/mean/max temperature per month')

plt.xlim( datetime(2015,10,1), datetime(2017,10,1) )
plt.xticks(rotation = 45)
plt.ylim(0.8*min(temp_min),1.2*max(temp_max))
```
